```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='placement_status', data=df)
plt.title('Placement Status Distribution')
plt.show()
```

### **Insights**

* More than half of the candidates successfully landed jobs, but a nearly equal and significant portion missed out on placement.

## Interpretation of your above chart

Explain the Chart.

Visualizing Success Rates: This chart is a univariate frequency bar chart that visualizes the categorical distribution of candidate placement outcomes across a massive dataset of roughly 100,000 records.

What is it saying?

A High but Competitive Placement Rate: The chart clearly shows that while the majority of individuals are successfully placed (approximately 54,000), a nearly equal, significant portion (approximately 46,000) remains unplaced.

What features you used?

Categorical Frequency Mapping: The visualization utilizes two primary features from the underlying dataset: the independent categorical variable placement_status on the x-axis ("Not Placed" vs. "Placed") and the aggregate count metric on the y-axis to represent the total volume of each group.

From this what you are showing us?

Almost nearly half of the students are not placed we need to find what is lacking them behind


```python
plt.figure(figsize=(10,5))
sns.countplot(y='branch', data=df, order=df['branch'].value_counts().index)
plt.title('Students by Branch')
plt.show()
```

## Interpretation of your above chart

Explain the Chart.

Visualizing Student Distribution: This is a horizontal bar chart that displays the total number of students enrolled across six different academic departments (branches) in a dataset of roughly 100,000 students.

What is it saying?

Heavy Dominance of Tech Branches: The chart clearly highlights a massive leaning toward software and tech fields, with CSE (Computer Science) dominating the student count at roughly 35,000, followed by IT at 20,000 and ECE at 15,000.

Equal Distribution in Core Fields: In contrast, the traditional core engineering branches—EEE, Mechanical, and Civil—share an identical, much lower enrollment footprint of exactly 10,000 students each.

What features you used?

Departmental Count Mapping: The visualization utilizes two main features: the independent categorical variable branch on the y-axis (representing the six academic disciplines) and the aggregate count metric on the x-axis to track the volume of students in each.

From this what you are showing us?

The dataset is imbalanced because CSE and IT have far more students than Civil or Mechanical.

This means future analysis, such as placement success by branch, may be dominated by CSE and IT results.

What is it saying?

Both placed and unplaced students share almost identical average (median) CGPAs of around 7.5, showing that grades alone do not guarantee placement.

What features you used?

The chart utilizes the continuous numerical variable cgpa on the y-axis and the categorical variable placement_status on the x-axis.

From this what you are showing us?

Iam  showing You that CGPA does not act as a strong standalone predictor for a student's placement status because the overall grade distributions are nearly matching.
